```python
from emot.emo_unicode import UNICODE_EMOJI, EMOTICONS_EMO
import re
import string 
# Pattern to match emojis
emoji_pattern = (
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
)

# Adding accented characters and special symbols to the allowed set
accented_characters = (
    u"\u00C0-\u00FF"  # Latin-1 Supplement (accented characters)
)

# Special symbols (UNICODE CHARS) to be allowed - important in specific contexts like legal, financial, scientific texts
special_symbols = '©®™±≠≤≥∞π∑§¶†•′″‰′′←→↑↓↔↕'


# Create the complete regex pattern string
pattern_string = (
    r'[^a-zA-Z0-9\s' + 
    ''.join([re.escape(x) for x in string.punctuation]) + 
    ''.join([re.escape(x) for x in currency_symbols]) + 
    special_symbols + 
    emoji_pattern + 
    accented_characters + 
    ']'
)

pattern = re.compile(pattern_string, re.UNICODE)
```

```python
import os
import re
import pandas as pd
from google.cloud import storage
from google.cloud import bigquery
from google.oauth2 import service_account
from tqdm import tqdm

# Service account key
key_path = '/Users/chkapsalis/Downloads/nlp-project-427710-3e1a48df3dba.json'
credentials = service_account.Credentials.from_service_account_file(key_path)

# Google Cloud project id
project_id = 'nlp-project-427710'
bucket_name = 'my-nlp-project-acg'

# Initialization of the BigQuery client
bigquery_client = bigquery.Client(project=project_id, credentials=credentials)
storage_client = storage.Client(project=project_id, credentials=credentials)

# Dataset and table information
dataset_id = 'covid'  # Replace with your dataset ID
table_id = 'covid1'  # Replace with your table ID

# Create the dataset if it does not exist
dataset_ref = bigquery_client.dataset(dataset_id)
dataset = bigquery.Dataset(dataset_ref)

try:
    bigquery_client.get_dataset(dataset_ref)  # Make an API request.
    print(f"Dataset {dataset_id} already exists.")
except:
    dataset = bigquery_client.create_dataset(dataset)  # Make an API request.
    print(f"Dataset {dataset_id} created.")

# Define the job configuration for loading data
job_config = bigquery.LoadJobConfig(
    schema=[
        bigquery.SchemaField("status_id", "INTEGER"),
        bigquery.SchemaField("user_id", "INTEGER"),
        bigquery.SchemaField("screen_name", "STRING"),
        bigquery.SchemaField("text", "STRING"),
        bigquery.SchemaField("source", "STRING"),
        bigquery.SchemaField("reply_to_status_id", "DECIMAL"),
        bigquery.SchemaField("reply_to_user_id", "DECIMAL"),  
        bigquery.SchemaField("reply_to_screen_name", "STRING"),
        bigquery.SchemaField("is_quote", "BOOL"),
        bigquery.SchemaField("is_retweet", "BOOL"),
        bigquery.SchemaField("favourites_count", "INTEGER"),
        bigquery.SchemaField("retweet_count", "INTEGER"),
        bigquery.SchemaField("country_code", "STRING"),
        bigquery.SchemaField("place_full_name", "STRING"),
        bigquery.SchemaField("place_type", "STRING"),
        bigquery.SchemaField("followers_count", "INTEGER"),
        bigquery.SchemaField("friends_count", "INTEGER"),
        bigquery.SchemaField("account_lang", "STRING"),
        #bigquery.SchemaField("account_created_at", ""),
        bigquery.SchemaField("verified", "BOOL"),
        bigquery.SchemaField("lang", "STRING")

    ],
    source_format=bigquery.SourceFormat.CSV,
    skip_leading_rows=1,
    field_delimiter='|',  # used this custom delimiter to help make sense out of the data
    autodetect=False,  # Automatically detect the schema
    max_bad_records=2000,  # Allow up to 2000 bad records
    ignore_unknown_values=True  # Ignore unknown values
)




import string 
def clean_dataframe(df):
    # Define a regular expression pattern to identify weird characters
    # explicitly getting rid of all quote characters because i have got errors of unclosed quotes
    # + I add emojis here, since I want to take them into account 
    # + I keep all currency symbols from the ISO standard (without taking into account historic denomincations - https://www.iso.org/iso-4217-currency-codes.html)
    # This will lead us to get rid of everything containing accent marks, but this is ok since we are only working with english texts for
    # everything except the machine translation part
    # !!! I do not drop single quotes because this way I would lose contractions !!! 
    # ~ later on, I will replace ` with ' to make sure noone has written contractions this way 
    # I have defined 'pattern' above


    # Filter out rows containing weird characters
    df_cleaned = df[~df.apply(lambda row: row.astype(str).str.contains(pattern).any(), axis=1)]

    # getting rid of double quotes because unclosed ones cause issues when loading the csv from the bucket into a bigquery table 
    df_cleaned['text'] = df_cleaned['text'].str.replace('"', ' ')
    
    # dropping escape characters from the remaining rows 
    # I do not drop '/' because it could break urls
    escaped_chars_pattern = re.compile(r'[' + ''.join([re.escape(c) for c in ['\\', '\t', '\n']])+ ']')
    df_cleaned['text'] = df_cleaned['text'].str.replace(escaped_chars_pattern, ' ', regex=True)

    # getting rid of excessive space characters in the remaining rows 
    excessive_whitespace_pattern = re.compile(r'\s{2,}')
    df_cleaned['text'] = df_cleaned['text'].str.replace(excessive_whitespace_pattern, ' ', regex=True)

    # getting rid of useless columns
    df_cleaned.drop(['created_at', 'account_created_at'], axis=1, inplace=True)

    # Remove the last character from column names and values because i got a trailing 'Â' char
    #df_cleaned.columns = [col[:-1] for col in df_cleaned.columns]
    #for column in df_cleaned.select_dtypes(include=['object']).columns:
    #    df_cleaned[column] = df_cleaned[column].apply(lambda x: x[:-1] if isinstance(x, str) else x)

    
    return df_cleaned


def process_and_clean_csv(input_path, output_path):
    # To clean the csv, I utilize the capabilities of pandas in making sence of the structure of csv files and then I 
    df = pd.read_csv(input_path, header=0, index_col=False, encoding='utf-8')
    
    # Clean the DataFrame
    df_cleaned = clean_dataframe(df)
    
    # Save the cleaned DataFrame to a new CSV file
    # this separator in the out file to make sure that I will be able to recognize the ending of cells
    df_cleaned.to_csv(output_path, index=False, encoding='utf-8', sep='|')

def upload_file(bucket_name, source_file_path, destination_blob_name):
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    
    # Set up the size of chunks to be uploaded and the timeout
    storage.blob._DEFAULT_CHUNKSIZE = 100 * 1024 * 1024  # 100 MB chunk size
    storage.blob._MAX_MULTIPART_SIZE = 100 * 1024 * 1024  # 100 MB max multipart size
    
    # Start the upload with increased timeout
    blob.upload_from_filename(source_file_path, timeout=600)  # 10 minutes timeout
    print(f'{source_file_path} uploaded to {bucket_name} as {destination_blob_name}.')

def load_to_bigquery(uri, table_ref, job_config):
    load_job = bigquery_client.load_table_from_uri(uri, table_ref, job_config=job_config)
    load_job.result()  # Wait for the load job to complete
    print(f'Loaded {uri} into {table_id}.')

source_folder = '/Users/chkapsalis/Downloads/covid1'
cleaned_folder = '/Users/chkapsalis/Downloads/cleaned'

# Ensure the cleaned folder exists
os.makedirs(cleaned_folder, exist_ok=True)

# Process each file in the source folder
for filename in tqdm(os.listdir(source_folder)):
    if filename.endswith('.csv'):
        source_file = os.path.join(source_folder, filename)
        cleaned_file = os.path.join(cleaned_folder, f'cleaned_{filename}')

        # Process and clean the CSV file
        process_and_clean_csv(source_file, cleaned_file)

        # Upload the cleaned file to GCS
        upload_file(bucket_name, cleaned_file, f'cleaned_{filename}')

        # Load the cleaned file into 
        uri = f'gs://{bucket_name}/cleaned_{filename}'
        load_to_bigquery(uri, dataset_ref.table(table_id), job_config)
```
